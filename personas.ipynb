## Set Personas

```python
from templates import SYSTEM_PROMPT

for p in personas.keys():
    # print(f"Persona ID: {p}")
    personas[p].update({"persona_name": " ".join(p.split('-')[2:]).title(),
                        "system_prompt": SYSTEM_PROMPT,
                        "pipeline_mode": "full",
                        "default_replica_id": os.getenv("REPLICA_ID")
                        })
    personas[p]['layers'].update({
        # "transport": {
        #     "transport_type": "livekit"
        # },
        "llm": {
            "model": "tavus-gpt-4o",
            # "temperature": 0.4,
            "speculative_inference": personas[p]['layers']['llm'].get('speculative_inference', True)
        },
        "stt": { "stt_engine": "tavus-advanced" },
        "tts": {
            "tts_engine": "elevenlabs",
            # "api_key": os.getenv("ELEVENLABS_API_KEY"),
            # "external_voice_id": "external-voice-id",
            # "tts_model_name": "sonic"
        }})
    personas[p].pop("default_replica_id")
    personas[p].pop("document_retrieval_strategy")
    personas[p].pop("memory")
    # personas[p]['speculative_inference'] = True

pp(personas['config-a-baseline-clinical-coach'])
```

## Create Personas (Try 1)

```python
import requests

url = "https://tavusapi.com/v2/personas"
results = []
for persona in personas.values():
    payload = persona
    headers = {
        "x-api-key": os.getenv("TAVUS_API_KEY"),
        "Content-Type": "application/json"
    }

    response = requests.post(url, json=payload, headers=headers)
    print(response.status_code)
    print(response.text)    
    results.append(response.json())
    print(f"Created persona {response.json().get('persona_id')} with name {persona['persona_name']}")

# Save results to a file
with open("created_personas.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=4)
```

## Create Personas (Try 2)

```python
import requests

url = "https://tavusapi.com/v2/personas"
results = []
for persona in personas.keys():
    payload = {
        "layers": {
            "transport": {
                "transport_type": "livekit"
            }
        },
        "persona_name": persona,
        "pipeline_mode": "echo"
    }
    headers = {
        "x-api-key": os.getenv("TAVUS_API_KEY"),
        "Content-Type": "application/json"
    }

    response = requests.post(url, json=payload, headers=headers)
    print(response.status_code)
    print(response.text)    
    results.append(response.json())
    print(f"Created persona {response.json().get('persona_id')} with name {persona}")

# Save results to a file
with open("created_personas.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=4)
```
